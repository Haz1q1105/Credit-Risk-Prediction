<p><b>Objective: Predicting whether a loan applicant is liekly to default on loan</p></b>
<p>Step 1: Importing all important libraries required</p>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split             #function for the training the data
from sklearn.preprocessing import Normalizer                     #normalizes data for model preparation
from sklearn.metrics import accuracy_score,confusion_matrix      #tests accuracy of the model
from sklearn.linear_model import LogisticRegression              #The library from where logistic regression will be imported
from sklearn.preprocessing import StandardScaler                 #library to scale our data in order to minimize biasness and confusion for the model

```

<h><b>Step 2: Importing Dataset and Performing Preprocessing and Cleaning</b></h>

```python
df = pd.read_csv("training_data.csv")
df         
```

```python
df.shape              #614 rows and 13 columns 
df.dtypes             #checking the data types of the features present in model
```

```python
df.describe()
```

```python
df.isnull().sum()      #checking if any null values present in the data
```

```python
#filling up missing values using .fillna() functions
df["Gender"] = df["Gender"].fillna("Not Specified")
df["Married"] = df["Married"].fillna("Not mentioned")
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])
df["Self_Employed"] = df["Self_Employed"].fillna("No mention")
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].mean())
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode()[0])
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].mode()[0])
```

```python
df.isnull().sum()      # re running the code to confirm if NaN values were successfully removed
```

<b>Step 3: Encoding categorical values for better training of the data </b>

```python
#encoding categorical labels for model to interpret and understand easily in binary because model can only understand binary or numericla values 
#encoding will ensure that cateegories are converted into boolean value into converted 
#selecting categorical columns and applying One-Hot Encoding 
df.dtypes             #verifying data types to identify categorical data
```

```python
#using drop_first to remove extra columns during the One-Hot Encoding
df_encoded = pd.get_dummies(df,drop_first = True)
```

```python
df_encoded.head(15)        #confirming if encoding is performed successfully 
```

```python
df_encoded.dtypes   #verifying boolean values 
```

```python
#since now all categorical columns have become boolean, we cannot use them in model training and testing. Therefore need to convert boolean values into numerical representation
b_cols = df_encoded.select_dtypes(include="bool").columns  #selects columns of type bool 
df_encoded[b_cols] = df_encoded[b_cols].astype(int)       #uses the selected boolean features and converts them into integer to represent True and False as 0 or 1
```

```python
df_encoded.dtypes      #the output will show boolean values converted into integers 
```

<b><h>Step 5: Model Training and Testing</h><b>
<p>Now model will be prepared for testing and training by using libraries from python</p>
<p>The following code will show the preparation for model training, with comments explaning the concept and reasoning behind each step</p>

```python
X = df_encoded.drop("Loan_Status_Y",axis=1)
Y = df_encoded["Loan_Status_Y"]

#step 0: importing important libaraies which already have been imported in the step 1 of the process
#------------------------------------------------------------------

#step 1: splitting data into training and testing data
#splitting data into training and testing 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=42,shuffle =True)
#random state : using random state to achieve consistent results and productive output
#test_size : splits the data. my model is using 80% of data to train and 20% to test 
#shuffle: allows model to be trained on different datasets and values by rearranging it to avoid biasness

#--------------------------------------------------------------------

#now scaling will be done on the dataset i prepared because if dataset includes values which are too big or too small, it can cause model 
#to behave in a biased way. Hence to prevent it we will scale it
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#--------------------------------------------------------------------
#using LogisticRegression function to implement binary regression
model = LogisticRegression(random_state=16)
#fitting the model with regression 
model.fit(X_train_scaled,Y_train)
y_pred = model.predict(X_test_scaled)
#model is prepared and ready for testing and running 
```

<b>Now running accuracy and Confusion Matrix</b>

```python
#the libraries for accraucy test and confusion matrix already loaded before

#checking accuracy:
acc = accuracy_score(Y_test,y_pred)
print("The model's accuracy score is: ",acc)
```

```python
#checking confusion matrix for better clarification and understanding
conf_matrix = confusion_matrix(Y_test,y_pred)
plt.figure(figsize=(7,5))
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap="Greens")
plt.title("Confusion Matrix",fontsize=16,weight='bold')
plt.xlabel("predicted")
plt.ylabel("Actual")
#                  ------------------------------------------------------------------------------
```
